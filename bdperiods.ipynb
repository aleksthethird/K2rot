```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf', 'jpg')
```

```python
year = np.arange(1998,2019)
nubd = np.array([0,1,1,5,7,10,20,30,35,40,40,150,170,170,170,180,180,200,200,200,260])
```

```python
%matplotlib inline
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
fig = plt.figure(figsize=(8, 6))
plt.xlim(1998,2019)
plt.ylim(0,300)
plt.xticks([2000,2005,2010,2015])
plt.xlabel('Year',fontsize=20)
plt.ylabel('~# brown dwarf periods',fontsize=20)
plt.plot(year,nubd,'-',linewidth=3)
```
